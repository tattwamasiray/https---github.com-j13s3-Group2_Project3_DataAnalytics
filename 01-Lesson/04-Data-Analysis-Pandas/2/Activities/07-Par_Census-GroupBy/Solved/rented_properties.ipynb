```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Save file path to variable
rentals_csv = "Resources/2016_rental_by_LGA.csv"
```

```python
# Read with Pandas
rentals_df = pd.read_csv(rentals_csv)
rentals_df.head()
```

```python
# Create a groupby object by state
state_total_group = rentals_df.groupby(["State"])
state_total_group
```

```python
# Create a dataframe of the totals for each state
state_totals_df = state_total_group.sum()
state_totals_df
```

```python
# Calculate the total rented properties in each state by summing all of the columns together.
# Place the results into a new column.
state_totals_df["Total"] = state_totals_df.sum(axis=1)

state_totals_df["Total"]
```

```python
# Create a dataframe of the averages for each state
state_avg_df = state_total_group.mean()
state_avg_df
```

```python
# Export the DataFrames to CSV
state_totals_df.to_csv("output/state_totals.csv", index=True)
state_avg_df.to_csv("output/state_avg.csv", index=True)
```
