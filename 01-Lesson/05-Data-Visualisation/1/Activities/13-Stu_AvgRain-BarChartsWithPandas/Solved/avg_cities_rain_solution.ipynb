```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
rain_df = pd.read_csv("../Resources/avg_rain_cities.csv")
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(8,5))
plt.bar(x_axis, rain_df["ANNUAL (mm)"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, rain_df["LOCATION"], rotation="vertical")
plt.show()
```

```python
# Set x and y limits
plt.xlim(-0.15, len(x_axis)-0.05)
plt.ylim(0, max(rain_df["ANNUAL (mm)"])+50)
plt.show()
```

```python
# Set a Title and labels
plt.title("Average Rain in Major Cities")
plt.xlabel("City")
plt.ylabel("Average Amount of Rainfall in Millimetres")
plt.show()
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("../Images/avg_cities_rain.png")
plt.show()
```
